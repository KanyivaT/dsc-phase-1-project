## Data Preperation

## Merging the Data

The purpose of merging data sets is to combine data from various sources into a single dataset, allowing for more comprehensive analysis, insights, and decision making.

```python
merged_data = pd.merge(title_ratings, title_basics, on='tconst', how='inner')
```

```python
final_concat = pd.concat([merged_data, movie_gross, reviews], axis=1)
final_concat.head()
```

## Data Cleaning

```python
# check for missing values
df =final_concat
df.isna().any()
```

```python
# The number of missing values under each column
df.isna().sum()
```

## Dropping Unnecessary Columns

```python
df = df.drop('start_year', axis=1)
```

```python
df = df.drop('primary_title', axis=1)
```

```python
df = df.drop('critic', axis=1)
```

```python
df = df.drop('top_critic', axis=1)
```

```python
df = df.drop('publisher', axis=1)
```

```python
df = df.drop('id', axis=1)
```

```python
df = df.drop('date', axis=1)
```

```python
df.isna().sum()
```

```python
# replacing the missing values with the mean
df['runtime_minutes'] = df['runtime_minutes'].fillna(value=df['runtime_minutes'].mean())
```

```python
# replacing the missing values with Unknown
df['genres'].fillna('Unknown Genre', inplace=True)
```

```python
df['title'].fillna('Unknown Title', inplace=True)
```

```python
# replacing the missing values with NaN
df['studio'].fillna('NaN', inplace=True)
```

```python
# replacing the missing values with a zero then changing the dtype from a float64 to an int64
df['domestic_gross'] = df['domestic_gross'].fillna(value=df['domestic_gross'].mean())
```

```python
# changing the dtype from an object to numeric(either float or integer)
# replacing the missing values with the mean so as not to lose the information
df['foreign_gross'] = pd.to_numeric(df['foreign_gross'], errors='coerce')
df['foreign_gross'] = df['foreign_gross'].fillna(value=df['foreign_gross'].mean())
```

```python
# replacing the missing values with No review
df['review'].fillna("No review", inplace=True)
```

```python
df['fresh'].fillna('Unknown', inplace=True)
```

```python
# Convert the non-numeric values to numeric values
# replacing the missing values with mean
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating'] = df['rating'].fillna(value=df['rating'].mean())
```

```python
df.dropna(subset=['year'], inplace=True)
```

## Analysis

```python
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# top 15 films with the highest average rating
df = df.sort_values(['averagerating'], ascending=False)
top_genres = df[ ['original_title', 'genres', 'averagerating'] ].head(15)
top_genres
```

```python
plt.figure(figsize=(12, 6))
plt.barh(top_genres['genres'], top_genres['averagerating'], color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Genre')
plt.title('Top Genres with Highest Average Ratings')  
plt.show()
```

```python
# top 15 films with the highest numvotes
df = df.sort_values(['numvotes'], ascending=False)
top_genres = df[ ['original_title', 'genres', 'numvotes'] ].head(15)
top_genres
```

```python
plt.figure(figsize=(12, 6))
plt.barh(top_genres['genres'], top_genres['numvotes'], color='orange')
plt.xlabel('Number of Votes')
plt.ylabel('Genre')
plt.title('Top Genres with Highest Number of Votes')  
plt.show()
```

```python
# top 15 genres based on the value counts
top_genres = df['genres'].value_counts().nlargest(15)
plt.figure(figsize=(12, 6))
plt.barh(top_genres.index, top_genres.values, color='yellow')
plt.title('Top 15 Genres Based on the Value Counts')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

```python
# top 15 genres based on the runtime
genre_runtimes = df.groupby('genres')['runtime_minutes'].mean()
top_15_genres = genre_runtimes.sort_values(ascending=False).head(15)
plt.figure(figsize=(12, 8))
top_15_genres.plot(kind='bar', color='purple')
plt.title('Average Runtime by Top 15 Genres')
plt.xlabel('Genre')
plt.ylabel('Average Runtime minutes')
plt.show()
```
